```python
import matplotlib.pyplot as plt
import scipy.ndimage as ndimage
from skimage import feature
from skimage import filters
import numpy as np
import imageio
%matplotlib inline
plt.rcParams.update({'image.interpolation': None})
plt.rcParams.update({'font.size': 15})
```

```python
I_easy = np.asarray(imageio.imread('easy1.jpg'))
Ir_easy = I_easy[:,:,0]
Ig_easy = I_easy[:,:,1]
Ib_easy = I_easy[:,:,2]
average_kernel = np.ones((3,3))/9
Ig_easy3 = ndimage.convolve(Ig_easy,average_kernel, mode="nearest")
plt.figure(figsize=(15,10))
plt.imshow(Ig_easy3,cmap="gray")
```

```python
I_easy = np.asarray(imageio.imread('easy1.jpg'))
Ir_easy = I_easy[:,:,0]
Ig_easy = I_easy[:,:,1]
Ib_easy = I_easy[:,:,2]
average_kernel = np.ones((5,5))/25
Ig_easy11 = ndimage.convolve(Ig_easy,average_kernel, mode="nearest")
plt.figure(figsize=(15,10))
plt.imshow(Ig_easy11,cmap="gray")
```

```python
I=Ig_easy-Ig_easy11
plt.figure(figsize=(15,10))
plt.imshow(I,cmap="gray")
```

```python

I2=I/255
sobel_horiz = np.array([[-1,-2,-1],[0,0,0],[1,2,1]])
sobel_vert = np.array([[-1,0,1],[-2,0,2],[-1,0,1]])
gx = ndimage.convolve(I2, sobel_horiz, mode='reflect')
gy = ndimage.convolve(I2, sobel_vert, mode='reflect')
g = np.sqrt((gx**2+gy**2))
thresh=np.percentile(g,99.7)
g_thresh = g >= thresh
Ip = (I2.astype(float))*(g_thresh)
T2_otsu=filters.threshold_otsu(I2[g_thresh],nbins=256)
I2s_ostu = I2.astype(float)>T2_otsu
plt.figure(figsize=(20,10))
plt.imshow(I2s_ostu,cmap="gray")
plt.axis('off')

plt.title('Using the Ostu method on the product image')
```

```python
I2=np.where(I2s_ostu==True,255,0)
plt.figure(figsize=(20,10))
plt.imshow(I2,cmap="gray")
```
